```python
titanic=titanic.drop(["Cabin"],axis="columns")
```

```python
titanic.isnull().sum()
```

```python
titanic["Age"]=titanic["Age"].fillna(titanic["Age"].median())
```

```python
titanic["Embarked"].fillna(titanic["Embarked"].mode())
```

```python
titanic["Embarked"]=titanic["Embarked"].fillna(titanic["Embarked"].value_counts().idxmax())
```

```python
titanic.groupby(["Sex","Survived"])
```

```python
titanic.groupby(["Sex","Survived"])["PassengerId"].count()
```

```python
titanic.groupby(["Pclass","Survived"])["PassengerId"].count()
```
